## Modifying Data within DataFrames
**apply:** It is used when you want to apply a function along the axis of a dataframe, it accepts a Series whose index is either column (axis=0) or row (axis=1). <br>For example: `df.apply(np.square)`, it will give a dataframe with number squared.  

**map:** It can be used only for a Series object and helps to substitutes the series value from the lookup dictionary, Series or a function and missing value will be substituted as NaN. Since it works only with series or dictionary so you can expect a better and optimized performance. <br>Example: `df[‘Col1’].map({‘Trenton’:’New Jersey’, ‘NYC’:’New York’, ‘Los Angeles’:’California’})`.  


**applymap:** It is used for element wise operation across one or more rows and columns of a dataframe. It has been optimized and some cases work faster than *apply* but it’s good to compare it with apply before going for any heavier operation. <br>Example: `df.applymap(np.square)`, it will give a dataframe with number squared.  

**replace:** Values of the DataFrame are replaced with other values dynamically. This differs from updating with `.loc` or `.iloc`, which require you to specify a location to update with some value.  

More on this <a href="https://kanoki.org/2019/11/25/pandas-apply-map-and-applymap/">here</a> and <a href="https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.replace.html">here</a>

```python
# Lets say we want to rename a column name
df.rename(columns={"ConvertedComp" : "SalaryUSD"})    # use inplace=True for actually applying this change
```

```python
# Lets say we want to replace the values of 'Hobbyist' column with Trues False instead of Yes No
df['Hobbyist'].map({'Yes' : True, 'No' : False})    # this method does not support inplace. assign it back to df['Hobbyist'] for preserving changes.
```

Note that if the column has values other than the keys specified in the dictionary passed to `map()` call, then those values are simply assigned to `NaN`  
To avoid modifying values other than specified keys, use `replace` function.

## Grouping and Aggregates
An essential piece of analysis of large data is efficient summarization; computing aggregations like `sum(), mean(), median(), min(),` and `max()`, in which a single number gives insight into the nature of a potentially large dataset.  

Simple aggregations can give you a flavor of your dataset, but often we would prefer to aggregate conditionally on some label or index. This is implemented in the so-called `groupby` operation. The name "group by" comes from a command in the SQL database language, but it is perhaps more illuminative to think of it in the terms of: ***split, apply, combine.***  

This makes clear what the groupby accomplishes:
1. The *split* step involves breaking up and grouping a DataFrame depending on the value of the specified key.
2. The *apply* step involves computing some function, usually an aggregate, transformation, or filtering, within the individual groups.
3. The *combine* step merges the results of these operations into an output array.



```python
# Lets find the median of salaries
df['ConvertedComp'].median()
```

```python
# describe() will automatically find numeric columns in our dataset and gives details of few of the aggregate functions 
df.describe()
```

```python
# Lets use groupby to group the dataframe based on countries
country_grp = df.groupby(['Country'])

# you can retrieve the details of particular country like this
country_grp.get_group('India')
```

**Example Query to understand grouping and aggregations:** Lets say for respondents of each country, we want to find out the percentage of those respondents who works with Python.

1. First lets group the respondents by country and calculate the number of respondents in each group:

```python
country_grp = df.groupby(['Country'])
country_respondents = df['Country'].value_counts()
```

2. Now, lets find out how many of them use Python and create a dataframe of number of respondents and number of them who know python:

```python
# finding out the sum of respondents who use Python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())

# creating a dataframe for number of respondents and those who know python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)

# renaming the column names for better understanding
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)

python_df
```

3. Finally, lets calculate the percentage of people who use Python

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100

# sort them based on their percentage in descending order
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)

python_df.head(50)
```

## Cleaning Data
**Dropping invalid values:**  
You can use `dropna()` to remove all the rows consisting `NaN` or `None` values in the dataframe. With `dropna()` you can specify  
- axis to delete from, using `axis` attribute
- specify conditions like 'drop row if any of the value in the row is `NaN`' or 'drop only if all the values of the row are `NaN`' by setting `how` attribute to `any` or `all`
- specify which columns to check for invalid values using `subset` attribute  

Sometimes, the dataset we use may represent invalid values with different notation such as 'Missing' or 'NA'.  
In such cases it is best practice to replace them with standard 'NAN' or 'None' values using `replace` method
Example:  
`
df.replace('NA', np.nan, inplace=True)  
df.replace('Missing', np.nan, inplace=True)`  

Coming to our 'Stack Overflow' dataset, it is better to replace these values while laoding the data from the csv file itself like this:  
`na_vals = ['NA', 'Missing']
df = pd.read_csv('data/survey_results_public.csv', index_col='Respondent', na_values=na_vals)`

You can also use `df.isna()` to check which values are invalid (returns dataframe filled with `True` or `False` values.  
You can use `fillna()` method to fill `NaN` or `None` values with some custom value like `0`. 


**Taking care of datatypes:**  
Sometimes, when you want to use aggregate functions like `mean()` or `sum()` on an numeric column values, it may fail because column contains `NaN` values (which is by default a float type value).  
So you can use `astype()` function to convert it to required dataype and carry out the required operation.  
Example: `df['age'] = df['age'].astype(float)`  



```python
# lets say we want to find out the average of YearsCoded column of our dataset.
# lets look at the values it contains
df['YearsCode'].unique()
```

```python
# we can see that it has couple of string values. lets clean that data
df['YearsCode'].replace('Less than 1 year', 0, inplace=True)
df['YearsCode'].replace('More than 50 years', 50, inplace=True)
# and also convert it to float to manage NaN values
df['YearsCode'] = df['YearsCode'].astype('float')
```

```python
# now finally, lets find out mean
df['YearsCode'].mean()
```

## Working with Dates and Time Series Data
Since the StackOverflow dataset doesn't contain Dates and Times values to work with, lets use another dataset which contains Etherium bitcoin price data for around two and a half years.  
(Dataset can be found <a href="https://github.com/CoreyMSchafer/code_snippets/blob/master/Python/Pandas/10-Datetime-Timeseries/ETH_1h.csv">here</a>)

```python
df2 = pd.read_csv('data/ETH_1h.csv')
df2.head()
```

The `Date` column here contains date and time values. To use methods like `day()` But it is not in the format of pandas datetime object. So we need to parse it to pandas supported format.

```python
df2['Date'] = pd.to_datetime(df2['Date'], format="%Y-%m-%d %I-%p")
df2['Date'].head()
```

```python
# now we can use required function on datetime object
df2.loc[0, 'Date'].day_name()
```

The better practice is to parse the data into appropriate format while loading the data from csv itself:

```python
d_parser = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %I-%p')
df2 = pd.read_csv('data/ETH_1h.csv', parse_dates=['Date'], date_parser=d_parser)
df2.head()
```

```python
# we can also use functions like
# find min data
df2['Date'].min()
```

```python
# find max date
df2['Date'].max()
```

```python
# find delta between two dates
df2['Date'].max() - df2['Date'].min()
```

```python
# retrieve data for specific timeframe
filt = (df2['Date'] >= pd.to_datetime('2019-01-01')) & (df2['Date'] < pd.to_datetime('2020-01-01'))
df2.loc[filt].head()
```

```python
# lets set the date as the index of the dataframe
df2.set_index('Date', inplace=True)
```

```python
# now we can easily access items via date indexing
# retreive all the data for year 2019
df2['2019']
```

```python
# retreive data between specific timeframe
df2['2020-01':'2020-02']
```

```python
# lets get the mean of closing price between Jan to Feb 2020 
df2['2020-01':'2020-02']['Close'].mean()
```

```python
# lets get the max value of highest price on Jan 1st 2020
df2['2020-01-01']['High'].max()
```

The above result is the maximum of all the values in dataset (resampled by hour by default).  
Now if we want the highs on daily or weekly basis, we have to resample it accordingly using the DateTime offset object.  
(You can find the codes <a href="https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects">here</a>)

```python
# lets resample on daily basis and find the high on a particular day
highs = df2['High'].resample('D').max()
highs['2020-01-01']
```
