### Lets plot a pie chart of most popularly used programming languages

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
# labels for each pie
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
# just for highlighting specific language (here - python)
explode = [0, 0, 0, 0.1, 0]

# shadow - adds a thin shadow around our pie chart (optional)
# autopct - calculated and adds percentage of each pie (recommended)
# startangle - rorate chart by certain degree (optional)
# wedgeprops - add an edge around the chart (optional)
plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

Note that the slices are displayed in the order of definition
