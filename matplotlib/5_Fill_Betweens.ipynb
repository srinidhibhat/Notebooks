### Lets plot the salaries of Python and Javascript developers in comparison with salaries of all developers based on the age and find out above average and below average salaries using 'fill betweens'. 

```python
import pandas as pd
from matplotlib import pyplot as plt

# lets read in the data 
# the data contains salaries of all developers, python developers and js developers for each age group
data = pd.read_csv('data/5_fill_betweens_data.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']

# lets plot the salaries of all developers in dotted line style
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

# now lets plot salaries of python developers in another line
plt.plot(ages, py_salaries, label='Python')

# alpha - transparency of fill between (optional)
# interpolate - making sure that certain x intersections are not clipped from plot
# lets fill between the above average salaries of python developers
plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries),
                 interpolate=True, alpha=0.25, label='Above Avg')

# lets fill between the below average salaries of python developers
plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

Looks like salaries of almost all python developers are above average (if you look closely between 35 to 40 age group, you can see it drops below average).

Lets plot the same for javascript developers

```python
js_salaries = data['JavaScript']

# lets plot the salaries of all developers in dotted line style
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

# now lets plot salaries of python developers in another line
plt.plot(ages, js_salaries, label='Javascript')

# lets fill between the above average salaries of javascript developers
plt.fill_between(ages, js_salaries, dev_salaries,
                 where=(js_salaries > dev_salaries),
                 interpolate=True, alpha=0.25, label='Above Avg')

# lets fill between the below average salaries of javascript developers
plt.fill_between(ages, js_salaries, dev_salaries,
                 where=(js_salaries <= dev_salaries),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
