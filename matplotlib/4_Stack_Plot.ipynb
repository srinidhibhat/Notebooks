### Lets plot number of work hours put in by employees each day on a certain project

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


day = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# this data corresponds to number of hours of work done in a certain project by each employee
# note that hours add up to 8 for each day
employee1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
employee2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
employee3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

labels = ['employee1', 'employee2', 'employee3']
colors = ['#6d904f', '#fc4f30', '#008fd5']

plt.stackplot(day, employee1, employee2, employee3, 
              labels=labels, colors=colors)

# just giving a location to legend such that it doesn't overlap the visible data
plt.legend(loc=(0.07, 0.05))

plt.title("Stack Plot")
plt.tight_layout()
plt.show()
```

This gives a nice visual representation of contribution of each employee to a particular project

Note that without labels specified for plot, legend cannot display the which color or part of the data corresponds to what
