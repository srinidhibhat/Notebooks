### Lets plot most commonly known programming languages in a horizontal bar chart

```python
import csv
import numpy as np
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

# read in the data using pandas csv_reader
data = pd.read_csv('data/1_bar_charts_data.csv')
ids = data['Responder_id']
# here multiple languages are separated by ';'
lang_responses = data['LanguagesWorkedWith']

# we use counter from collections for keeping a count of languages known by everyone
language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))

languages = []
popularity = []

# lets store only top 10 most known languages for plotting
for item in language_counter.most_common(10):
    languages.append(item[0])
    popularity.append(item[1])
    
# lets plot them in decreasing order of popularity
languages.reverse()
popularity.reverse()

# horizontal bar chart
plt.barh(languages, popularity)

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")

# add some padding
plt.tight_layout()

plt.show()
```
